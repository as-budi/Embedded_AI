```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset Iris
iris = load_iris()
X = iris.data  # Data fitur
y = iris.target  # Label kelas

# Buat DataFrame dari fitur dan label
df = pd.DataFrame(X, columns=iris.feature_names)  # Menambahkan nama kolom
df['target'] = y  # Menambahkan kolom target
df['target_name'] = df['target'].map({i: name for i, name in enumerate(iris.target_names)})  # Tambahkan label kelas

# Simpan ke file Excel
df.to_excel("iris_dataset.xlsx", index=False)

print("Dataset Iris berhasil diekspor ke iris_dataset.xlsx")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Load dataset Iris
iris = load_iris()
X = iris.data  # Data fitur (4 dimensi)
y = iris.target  # Label kelas

print ("Data fitur (X):")
print (X)
print ("Label kelas (y):")
print (y)

# Lakukan PCA untuk mereduksi ke 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot hasil reduksi dimensi
plt.figure(figsize=(8, 6))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k', alpha=0.7)
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA: Reduksi Dimensi dari 4D ke 2D pada Dataset Iris")
# plt.legend(handles=scatter.legend_elements()[0], labels=iris.target_names)
plt.grid(True)
plt.show()
```
