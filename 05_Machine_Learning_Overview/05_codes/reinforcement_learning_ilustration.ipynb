```python
import numpy as np
import matplotlib.pyplot as plt

# Simulasi lingkungan grid sederhana untuk Reinforcement Learning
grid_size = 5
goal_state = (4, 4)
obstacle_states = [(2, 2), (3, 3)]
agent_path = [(0, 0), (1, 0), (1, 1), (2, 1), (3, 1), (3, 2), (4, 2), (4, 3), (4, 4)]

# Plot grid environment
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xticks(np.arange(grid_size + 1) - 0.5, minor=True)
ax.set_yticks(np.arange(grid_size + 1) - 0.5, minor=True)
ax.grid(which="minor", color="black", linestyle="-", linewidth=1)
ax.tick_params(which="both", bottom=False, left=False, labelbottom=False, labelleft=False)

# Mark goal state
ax.add_patch(plt.Rectangle(goal_state, 1, 1, color="green", alpha=0.5, label="Goal"))

# Mark obstacles
for obs in obstacle_states:
    ax.add_patch(plt.Rectangle(obs, 1, 1, color="red", alpha=0.5, label="Obstacle" if obs == obstacle_states[0] else ""))

# Plot agent path
agent_x, agent_y = zip(*agent_path)
ax.plot(np.array(agent_x) + 0.5, np.array(agent_y) + 0.5, "bo-", markersize=8, label="Agent Path")

# Mark start position
ax.add_patch(plt.Circle(agent_path[0], 0.3, color="blue", label="Start"))

# Legend
ax.legend(loc="upper left", bbox_to_anchor=(1, 1))

plt.title("Reinforcement Learning: Agent Navigation in Grid World")
plt.show()
```
