## **1. Metode Normalisasi**
### **A. Min-Max Scaling**
$
X_{\text{normalized}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$
- Hasilnya dalam rentang **0 hingga 1**.
- Cocok untuk model berbasis **neural network** atau algoritma yang sensitif terhadap skala fitur.

### **B. Z-Score Standardization**
$
X_{\text{standardized}} = \frac{X - \mu}{\sigma}
$
- Hasilnya memiliki **mean = 0** dan **standar deviasi = 1**.
- Cocok untuk algoritma yang **membutuhkan distribusi Gaussian**, seperti **KNN, PCA, atau regresi linear**.

---

## **3. Analisis Hasil**
- **Grafik 1 (Merah)**: Data suhu asli dengan noise.
- **Grafik 2 (Biru)**: **Min-Max Scaling**, semua nilai berada dalam rentang **0 hingga 1**.
- **Grafik 3 (Hijau)**: **Z-Score Standardization**, mean **0** dan standar deviasi **1**.

**Kesimpulan**:
- **Gunakan Min-Max Scaling** jika algoritma yang digunakan sensitif terhadap skala fitur.
- **Gunakan Z-Score Standardization** jika data perlu didistribusikan secara normal.

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Simulasi data suhu (seperti pada contoh sebelumnya)
np.random.seed(42)
time = np.linspace(0, 10, 500)  # 10 detik, 500 sampel
true_temperature = 25 + 2 * np.sin(2 * np.pi * 0.1 * time)  # Suhu sebenarnya
sensor_temperature = true_temperature + np.random.normal(0, 1, size=len(time))  # Noise 1C

# 2. Min-Max Scaling
X_min = np.min(sensor_temperature)
X_max = np.max(sensor_temperature)
sensor_temperature_minmax = (sensor_temperature - X_min) / (X_max - X_min)

# 3. Z-Score Standardization
X_mean = np.mean(sensor_temperature)
X_std = np.std(sensor_temperature)
sensor_temperature_zscore = (sensor_temperature - X_mean) / X_std

# 4. Visualisasi hasil normalisasi
plt.figure(figsize=(12, 6))

plt.subplot(3, 1, 1)
plt.plot(time, sensor_temperature, 'r-', label="Original Sensor Data")
plt.title("Original Sensor Data (Temperature)")
plt.ylabel("Temperature (C)")
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(time, sensor_temperature_minmax, 'b-', label="Min-Max Normalized")
plt.title("Min-Max Normalization (Range 0 to 1)")
plt.ylabel("Normalized Value")
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(time, sensor_temperature_zscore, 'g-', label="Z-Score Standardized")
plt.title("Z-Score Standardization (Mean=0, Std=1)")
plt.ylabel("Standardized Value")
plt.xlabel("Time (s)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
